```python
import pandas as pd
import numpy as np
```

# Data Cleaning

```python
df = pd.read_csv('hello.dat', header=None)
df.head(3)
```

```python
dataframe = df.rename(columns = {0: 'nuevo_titulo'})
dataframe.head(3)
```

```python
# Dividir 'nuevo_titulo' en las tres primeras columnas
dataframe['fips'] = dataframe['nuevo_titulo'].str[:2]
dataframe['id'] = dataframe['nuevo_titulo'].str[:11]
dataframe['year'] = dataframe['nuevo_titulo'].str[11:15]
dataframe['element'] = dataframe['nuevo_titulo'].str[15:19]
dataframe['jan'] = dataframe['nuevo_titulo'].str[20:24]
dataframe['feb'] = dataframe['nuevo_titulo'].str[28:32]
dataframe['mar'] = dataframe['nuevo_titulo'].str[36:40]
dataframe['apr'] = dataframe['nuevo_titulo'].str[44:48]
dataframe['may'] = dataframe['nuevo_titulo'].str[52:56]
dataframe['jun'] = dataframe['nuevo_titulo'].str[60:64]
dataframe['jul'] = dataframe['nuevo_titulo'].str[68:72]
dataframe['aug'] = dataframe['nuevo_titulo'].str[76:80]
dataframe['sep'] = dataframe['nuevo_titulo'].str[84:88]
dataframe['oct'] = dataframe['nuevo_titulo'].str[92:96]
dataframe['nov'] = dataframe['nuevo_titulo'].str[100:104]
dataframe['dec'] = dataframe['nuevo_titulo'].str[108:112]
dataframe.head(3)
```

```python
df = dataframe.drop('nuevo_titulo', axis=1)
df.head(3)
```

```python
df[['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']] = df[['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']].astype(float)
```

```python
df.head()
```

```python
def division(cell):
    if cell.dtype == 'float64':
        return cell / 100
    else:
        return cell
```

```python
df = df.apply(division)
```

```python
countries = pd.read_fwf('countries.txt', header= None)
countries
```

```python
countries = countries.rename(columns = {0: 'fips', 1: 'country'})
new_row = {'fips': 'FO', 'country': 'Faroe Islands'}
countries = countries.append(new_row, ignore_index=True)
countries
```

```python
merge = pd.merge(df, countries, on='fips', how='outer')
merge.head(3)
```

```python
num_filas = len(merge)
nuevas_filas = num_filas - 6
df_merge = merge.head(nuevas_filas)
df_merge
```

```python
df_merge_copy = df_merge.copy()
```

```python
filas_con_nulos = df_merge_copy[df_merge.isnull().any(axis=1)]
filas_con_nulos
```

```python
for col in df_merge_copy.columns[1:]:
    for i in filas_con_nulos.index:
        if pd.isna(df_merge_copy.loc[i, col]):
            prev_val = df_merge_copy.loc[i-1, col]
            next_val = df_merge_copy.loc[i+1, col]
            j = 1
            while np.isnan(prev_val):
                j += 1
                prev_val = df_merge_copy.loc[i-j, col]
            k = 1
            while np.isnan(next_val):
                k += 1
                next_val = df_merge_copy.loc[i+k, col]
            mean_val = ((prev_val * 1/j) + (next_val * 1/k)) / (1/j + 1/k)
            df_merge.loc[i, col] = mean_val
```
